#EV Charging Station App - Richard Yanchev

The goal of this project is to simplify and streamline the charging process for electric vehicle owners by creating an app to organize location, specification, and price data of electric vehicle charging stations, all of which is stored in a database.

This phone and web app provides electric vehicle owners the ability to locate chargers on demand based on specific criteria such as company or brand, distance, price (ex. price per minute, price per watt, etc.), charging standards (ex. Type 1, Type 2, GB-T, CHADEMO, Tesla SC, etc.), charging speed, and customer ratings (to indicate whether the charging experience was pleasant or troublesome). A customer would want access to such information to help with finding a charger that is compatible with their vehicle, charges quickly or slowly, is expensive or inexpensive, and how far away each charging station is. Other vital information would include whether a charging station is in use or whether it is in operation (as it may be broken or under maintenance). Like a gas station, these charging stations will be almost constantly in use with customers needing to re-charge their electric vehicles frequently. It is important to keep an up-to-date database of charging stations to streamline the "refueling" or charging process for electric vehicle owners.
